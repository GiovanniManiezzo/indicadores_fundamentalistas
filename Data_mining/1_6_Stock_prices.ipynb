```python
import yfinance as yf
import pandas as pd
import numpy as np
import os
from dotenv import load_dotenv
from yahooquery import search
```

```python
load_dotenv()

#Pesquisar preços das ações de empresas de transmissão de energia elétrica
#* ELET4
#* ISAE4
#* TAEE4
#* ALUP4
ticker_acoes = ['ELET6.SA', 'ISAE4.SA', 'TAEE4.SA', 'ALUP4.SA']
valores = {}

for ticker in ticker_acoes:
    tickerData = yf.Ticker(ticker)
    tickerDf = tickerData.history(period="3mo", start='2010-03-01', end='2024-10-31') 
    valores[ticker] = tickerDf['Close']
    print(ticker + ' finalizado')
```

```python
valores

#transformar valores em um dataframe
df = pd.DataFrame(valores)
df.info()
```

```python
df = df.rename_axis('data')
df.index = df.index.tz_localize(None)
df = df.resample('QE').last()
df = df.fillna(0)
```

```python
df.tail(50)
```

```python
path_dados_tratados = os.getenv('PATH_DADOS_TRATADOS')
df.to_csv(path_dados_tratados + 'preco_acoes.csv')
```
